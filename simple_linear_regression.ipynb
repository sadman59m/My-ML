# Simple Linear Regression
- Uses the eqn y = b0 + b1x1
- y is the dependent value, x is the indepentend value

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# importing datase and splitting to X and y

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

Split into Train Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
```

# Training the simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# Predict The Result

```python
y_pred = regressor.predict(X_test)
```

Visualizing the Training Set Result

```python
# plot the train set as dots
plt.scatter(X_train, y_train, color = 'red')
# plot the result which is a line
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

# Visualizing Test Set results

```python
# plot the train set as dots
plt.scatter(X_test, y_test, color = 'red')
# Keep the same plot, cz it's on the unique equation we get from training the model
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
